```python
data_dict = df.set_index("Business Title")["Salary Range From"].to_dict()
```

```python
data_dict
```

```python
def find_salary_by_title(data, title):
    return data.get(title, "Title not found")
```

```python
program_manager_salary = find_salary_by_title(data_dict, "Program Manager")
print(f"Salary for 'Program Manager': {program_manager_salary}")
```
