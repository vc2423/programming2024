```python
merged_df = pd.merge(df2022, df2020, how = "outer")
```

```python
merged_df
```

```python
merged_with_year_df = pd.merge(df2022, df2020, how="outer")
```

```python
merged_with_year_df
```

```python
merged_with_year_df[merged_with_year_df["Fruit"] == "Apples"].groupby("year").mean("RetailPrice")
```

```python
merged_with_year_df[merged_with_year_df["Fruit"] == "Strawberries"].groupby("year").mean("RetailPrice")
```

```python
merged_with_year_df[merged_with_year_df["Fruit"] == "Watermelon"].groupby("year").mean("RetailPrice")
```

```python
price_comparison = merged_with_year_df.groupby(['Fruit', 'RetailPrice']).agg(
    MinRetailPrice=('RetailPrice', 'min'),
    MaxRetailPrice=('RetailPrice', 'max'),
    AverageRetailPrice=('RetailPrice', 'mean'),
    CountEntries=('RetailPrice', 'count')
).reset_index()

print("Retail Price Comparison:")
print(price_comparison)
```
